# Which hosts are the busiest and why?

###### Sonder(NYC) has the most listings so they are busiest. They are also generally open a long time as well.

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.groupby(["calculated_host_listings_count", "host_name"], as_index=False).count().sort_values("calculated_host_listings_count", ascending=False)
```

# How many neighborhood groups are available and which shows up the most?

##### Manhattan

```python
air_bnb.groupby("neighbourhood_group").count()
```

# Are private rooms the most popular in manhattan?

##### Yes

```python
air_bnb.groupby(["neighbourhood_group", "room_type"]).count()
```

# Which hosts are the busiest and based on their reviews?

##### Maya is based on reviews

```python
air_bnb.groupby(["host_id", "host_name"]).sum()[["number_of_reviews"]].sort_values("number_of_reviews", ascending=False)
```

# Which neighorhood group has the highest average price?

##### Manhattan

```python
air_bnb.groupby("neighbourhood_group").mean()[["price"]]
```

# Which neighbor hood group has the highest total price?

##### Manhattan

```python
air_bnb.groupby("neighbourhood_group").sum()[["price"]]
```

# Which top 5 hosts have the highest total price?

##### Sonder (NYC)	82795
##### Blueground     70331
##### Sally                 37097
##### Red Awning     35294
##### Kara                  33581

```python
air_bnb.groupby(["host_id", "host_name"]).sum()[["price"]].sort_values("price", ascending=False)
```

# Who currently has no (zero) availability with a review count of 100 or more?

```python
air_bnb.groupby(["host_id", "host_name","availability_365"])
#air_bnb.groupby("availability_365")[["host_id", "host_name","availability_365", "number_of_reviews"]]
air_bnb[(air_bnb['number_of_reviews'] >=100) & (air_bnb['availability_365'] == 0)].groupby(["host_id", "host_name"]).sum()[["number_of_reviews","availability_365"]]

# for x in air_bnb:
#     if air_bnb["number_of_reviews"].any() > 100 & air_bnb["availability_365"].any() == 0:
#         display(air_bnb[["host_id", "host_name"]])

# for x in air_bnb["availability_365"]:
#     if x == 0:
#         print("bla")
```

# What host has the highest total of prices and where are they located?

##### Sonder (NYC)	Manhattan	82795

```python
air_bnb.groupby(["host_id", "host_name", "neighbourhood_group"]).sum()[["price"]].sort_values("price", ascending=False)
```

# When did Danielle from Queens last receive a review?
